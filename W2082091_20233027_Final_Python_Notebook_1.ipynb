Code Reuse Session : 01 | Tutorial No : 01(Page 35)

```python
#function to determine outliers
def find_outliers_IQR(data):
  q1=data.quantile(0.25)
  q3=data.quantile(0.75)
  IQR=q3-q1
  outliers = data[((data<(q1-1.5*IQR))|(data > (q3+1.5*IQR)))]
  return outliers
```

Code Reuse Session : 01 | Tutorial No : 01(Page 35)

```python
#detect and count outliers in each numeric column using the IQR method
for column in dataset.select_dtypes(include = 'number').columns:
  outliers = find_outliers_IQR(dataset[column])
  print("Number of outliers in " + column + " : " + str(len(outliers)))
```

Code Reuse Session : 01 | Tutorial No : 01(Page 37)

```python
#replace outlier values in the 'Age' column with the mean age using the IQR method
outliers_age = find_outliers_IQR(dataset['Age']).dropna().index
mean_age = dataset['Age'].mean()
dataset.loc[outliers_age, 'Age'] = mean_age

#fill missing values in the 'Sex' column using the mean age
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
```

Code Reuse Session : 01 | Tutorial No : 01(Page 37)

```python
#replace outlier values in the 'Tumor_Size' column with the mean tumor size using the IQR method
outliers = find_outliers_IQR(dataset['Tumor_Size']).dropna().index
mean_tumor_size = dataset['Tumor_Size'].mean()
dataset.loc[outliers, 'Tumor_Size'] = mean_tumor_size

#fill missing values in the 'tumor' column using the mean tumor size
dataset['Tumor_Size'] = dataset['Tumor_Size'].fillna(dataset['Tumor_Size'].mean())
```

Code Reuse Session : 01 | Tutorial No : 01(Page 37)

```python
#replace outlier values in the 'Regional_Node_Examined' column with the mean Regional Node Examined size using the IQR method
outliers = find_outliers_IQR(dataset['Regional_Node_Examined']).dropna().index
mean_regional_node_examined = dataset['Regional_Node_Examined'].mean()
dataset.loc[outliers, 'Regional_Node_Examined'] = mean_regional_node_examined

#fill missing values in the 'Regional_Node_Examined' column using the mean regional node examined
dataset['Regional_Node_Examined'] = dataset['Regional_Node_Examined'].fillna(dataset['Regional_Node_Examined'].mean())
```

Code Reuse Session : 01 | Tutorial No : 01(Page 37)

```python
#replace outlier values in the 'Regional_Node_Positive' column with the mean regional node positive using the IQR method
outliers = find_outliers_IQR(dataset['Regional_Node_Positive']).dropna().index
mean_regional_node_positive = (dataset['Regional_Node_Positive'].mean()).astype(int)
dataset.loc[outliers, 'Regional_Node_Positive'] = mean_regional_node_positive

#fill missing values in the 'Regional_Node_Positive' column using the mean regional node positive
dataset['Regional_Node_Positive'] = dataset['Regional_Node_Positive'].fillna(dataset['Regional_Node_Positive'].mean())
```

```python
#remove all rows containing outliers in any numeric column using the IQR method
outlier_index = set()
for column in dataset.select_dtypes(include = 'number').columns:
  outliers = find_outliers_IQR(dataset[column])
  outlier_index.update(outliers.dropna().index)
dataset.drop(index = outlier_index, inplace=True)
```

```python
#Convert 'Female' in the 'Sex' column to the numerical value
dataset['Sex'] = dataset['Sex'].replace('Female', 2)

#Convert categorical values in 'T_Stage' column to numeric using mapping
dataset['T_Stage'] = dataset['T_Stage'].map({'T1':1, 'T2':2, 'T3':3, 'T4':4})

#Convert categorical values in 'N_Stage' column to numeric using mapping
dataset['N_Stage'] = dataset['N_Stage'].map({'N1':1, 'N2':2, 'N3':3})

#Convert categorical values in '6th_Stage' column to numeric using mapping
dataset['6th_Stage'] = dataset['6th_Stage'].map({'IIA':1, 'IIIA':2, 'IIIC':3, 'IIB':4, 'IIIB':5})

#Convert categorical values in 'N_Stage' column to numeric using mapping
dataset['A_Stage'] = dataset['A_Stage'].map({'Regional' : 1 , 'Distant' : 2})

#Convert categorical values in 'Estrogen_Status' column to numeric using mapping
dataset['Estrogen_Status'] = dataset['Estrogen_Status'].map({'Positive': 1, 'Negative': 0})

#Convert categorical values in 'Progesterone_Status' column to numeric using mapping
dataset['Progesterone_Status'] = dataset['Progesterone_Status'].map({'Positive': 1, 'Negative': 0})

#Convert categorical values in 'Mortality_Status' column to numeric using mapping
dataset['Mortality_Status'] = dataset['Mortality_Status'].map({'Alive': 1, 'Dead': 0})
```
